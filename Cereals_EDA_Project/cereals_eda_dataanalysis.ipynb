```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('cereal.csv')
```

```python
df.head()
```

### Plot ratings of different types of manufacturers.

```python
plt.figure(figsize=(10,5))
sns.barplot(x='rating',y='mfr',data=df)
plt.title("Rating of different Manufacturers")
plt.xlabel("Rating")
plt.ylabel("Manufacturers")
plt.show()
```

### Use xticks range form 0-100.

```python
plt.figure(figsize=(10,5))
sns.barplot(x='rating',y='mfr',data=df)
plt.title("Rating of different Manufacturers")
plt.xlabel("Rating")
plt.ylabel("Manufacturers")

plt.xticks(range(0,101,10))

plt.show()
```

####  Change the style of the graph to seaborn 

```python
sns.set_style('whitegrid')
plt.figure(figsize=(10,5))
sns.barplot(x='rating',y='mfr',data=df)
plt.title("Rating of different Manufacturers")
plt.xlabel("Rating")
plt.ylabel("Manufacturers")

plt.xticks(range(0,101,10))

plt.show()
```

### Insights 
####  Manufacturers show variation in ratings , where some brands consistently performing better, indicating higher quality and preference.
