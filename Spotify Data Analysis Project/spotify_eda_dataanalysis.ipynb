```python
import pandas as pd
```

```python
df=pd.read_csv("top50spotify.csv")
```

```python
df.head()
```

### Drop 1st column

```python
df=df.drop(df.columns[0],axis=1)
df.head()
```

### Save this file as top50.csv


df.to_csv("top50.csv", index=False)


### Find the average Energy and Length of first 10 songs. 

```python
first_10=df.head(10)

avg_energy=first_10['Energy'].mean()

avg_length=first_10['Length.'].mean()
```

```python
print('average energy is',avg_energy)
print('average Length is',avg_length)
```

### Find the total length of songs, group by genre from top to bottom.

```python
df_length= df.groupby('Genre')['Length.'].sum()

# sorting values from top to bottom i.e from highest to lowest
df_length_sort=df['Length.'].sort_values(ascending=False)

print("Sorted list from top to bottom is: \n",df_length_sort)
```

### Print the artist name with the most number of tracks in one genre. (Hint: Group by artist name and genre) 

```python
artist_genre=df.groupby(['Artist.Name','Genre'])['Track.Name'].count().reset_index() # reset_index is used to see output in table format
print(artist_genre.sort_values('Track.Name',ascending=False).head(5))
```

```python
# above output is coming with index so using to_string to remove index
print(artist_genre.sort_values('Track.Name',ascending=False).head(1).to_string(index=False))
```

### Print the data of the tracks created by the artist from the previous question.

```python
artist_data=df.groupby(['Artist.Name','Genre'])['Track.Name'].count().reset_index()
artist_data=artist_data.sort_values('Track.Name',ascending=False)
name=artist_data.iloc[0]['Artist.Name']

print(df[df['Artist.Name']==name])
```
